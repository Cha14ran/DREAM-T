```
import pandas as pd
import matplotlib.pyplot as plt


#Sentiment Analysis three class



#only over-sampling
data1={'LEXICON_BASED':[57,51,54],'BOW':[54,54,49],'TF-IDF':[55,55,54]}
data2={'Word2vec':[59,62,61],'GloVe':[58,61,61],'FastText':[63,65,63],'Meta-Embeddings':[61,65,63]}
data3={'Skip-Thought':[58,57,57],'BERT':[67,67,67],'ELMo':[63,64,63]}

# data=data1+data2+data3
data2.update(data1)
data3.update(data2)
df=pd.DataFrame(data3)
# print("Sentiment Analysis Box-plot")
# boxplot1=df.boxplot(column=['LEXICON_BASED','BOW','TF-IDF','Word2vec','GloVe','FastText','Meta-Embeddings','Skip-Thought','BERT','ELMo'],grid=False,rot=45,figsize=(20,15),fontsize=18)
# plt.figtext(.5,.9,'Sentiment Analysis Box-plot', fontsize=50, ha='center')


boxprops = dict(linestyle='-', linewidth=3, color='k')
medianprops = dict(linestyle='-', linewidth=3, color='k')
bp=df.boxplot(column=['LEXICON_BASED','BOW','TF-IDF','Word2vec','GloVe','FastText','Meta-Embeddings','Skip-Thought','BERT','ELMo'],grid=False,rot=45,figsize=(20,15),fontsize=18,
              showfliers=True, showmeans=True,
              boxprops=boxprops, medianprops=medianprops,
              return_type='dict')

plt.ylabel("F1-Score",fontsize=20)
plt.suptitle('Sentiment Analysis(Three Class) Box-plot', fontsize=40)


plt.savefig("sentiment_3class.png")
```
